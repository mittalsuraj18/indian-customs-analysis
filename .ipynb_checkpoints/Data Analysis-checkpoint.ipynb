```python
def chisq_of_df_cols(df, c1, c2):
    groupsizes = df.groupby([c1, c2]).size()
    ctsum = groupsizes.unstack(c1)
    # fillna(0) is necessary to remove any NAs which will cause exceptions
    return(chi_square(ctsum.fillna(0)))
```

```python
columns = list(data.columns)
columns.remove('value_of_goods_in_rupees')
columns.remove('description_of_goods')
columns
```

```python
chi_square_data = []
for i in columns:
    temp = []
    for j in columns:
        if j is not i:
            display(i,j)
            chi,_,_,_ = chisq_of_df_cols(data,i,j)
        else:
            chi = 1
        temp.append(chi)
    chi_square_data.append(temp)
```

```python
trace = Heatmap(z=chi_square_data,
                   x=columns,
                   y=columns,)
heatmap_data=[trace]
plt_offline.iplot(heatmap_data, filename='labelled-heatmap')
```
